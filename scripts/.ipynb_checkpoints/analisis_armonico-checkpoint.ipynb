# Me fijo si la serie tiene una tendencia significativa

```python
import pymannkendall as mk

"""
And all Mann-Kendall tests return a named tuple which contained:

trend: tells the trend (increasing, decreasing or no trend)
h: True (if trend is present) or False (if the trend is absence)
p: p-value of the significance test
z: normalized test statistics
Tau: Kendall Tau
s: Mann-Kendal's score
var_s: Variance S
slope: Theil-Sen estimator/slope
intercept: intercept of Kendall-Theil Robust Line, for seasonal test, full period cycle consider as unit time step
sen's slope function required data vector. seasonal sen's slope also has optional input period, which by the default value is 12. Both sen's slope function return only slope value.
"""

#Hago un test de Mann kendall para ver si tiene una parte determinstica.

alpha_mk=0.05
resultado_mk = mk.original_test(santotome.altura_rio, alpha=alpha_mk)
print(f'se concluye que se rechaza la hipotesis nula, en favor de la alternativa con una sifnificancia del 5%: Tendencia = {resultado_mk[1]} con alpha de {alpha_mk}')
```

```python
ax1 = plt.subplot(1,2,1)
sgt.plot_acf(santotome.altura_rio, lags = len(santotome.altura_rio)*0.3, ax=ax1)
plt.title('Autocorrelograma de altura de ro - Santo Tom (30% de lags)', size=20)
plt.xlabel('n de lags')
plt.ylabel('valor de correlacin')
plt.grid()
ax2=plt.subplot(1,2,2)
sgt.plot_acf(santotome.altura_rio, lags = 150, ax=ax2).set_size_inches(40,10)
plt.title('Autocorrelograma de altura de ro - Santo Tom (100 lags)', size=20)
plt.xlabel('n de lags')
plt.ylabel('valor de correlacin')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
rl = stats.linregress(y=santotome.altura_rio, x=np.arange(len(santotome)), alternative='two-sided')
```

```python
rl
```

```python
x = np.arange(len(santotome))
y = rl[0]*x + rl[1]
```

```python
regresion_lineal = pd.DataFrame({'rl':y}, index=santotome.index)
```

```python
plt.figure(figsize=(20,10))

plt.plot(santotome, c='b', label='altura de rio')
plt.plot(regresion_lineal, c='r', label=f'regresion lineal: y={rl[0]:.6f} x + {rl[1]:.6f}')
plt.xlabel('fecha')
plt.ylabel('altura de rio [m]')
plt.grid()
plt.legend()
```

### Resultado del test de pendiente:

La recta resultante de la regresin lineal por mnimos cuadrados es: y = x * 0.00015829346164588192 + 3.3354802186345864. Y testeando con un test de Wald, cuya hiptesis nula es que el valor de la pendiente es 0 (cero), con un 5% de significancia. 

El p-valor es de: 2.6700063343165473e-51, por lo que se rechaza la hiptesis nula en favor de la alternativa, que la pendiente es distinta de cero. (two-sided: the slope of the regression line is nonzero).  
Como no es nula la tendencia, la filtro para poder realizar un anlisis armnico.

```python
santotome_sin_tendencia = santotome.altura_rio - regresion_lineal.rl
```

```python
plt.figure(figsize=(20,10))
santotome_sin_tendencia.plot()
plt.grid()
plt.title('Altura de ro sin tendencia para la estacin de santotome', size=20)
plt.xlabel('fecha')
plt.ylabel('altura de ro [m]')
```

```python
valor = santotome_sin_tendencia[:-1]
N  =  len(valor)
K = N/2
P = N	

########CAMBIAR EL N?MERO DE ARMONICO
M = 8 #ESTE VALOR HAY QUE CAMBIAR (el armnico que quiero filtrar)

###       Calcula el promedio

PROM = valor.mean()  

PI = 3.1415926

########### comienza!
##busco los coef A y B del armnico a filtrar

SUM = 0
SAM = 0

for J in np.arange(N): 
    SUM = SUM+valor[J]*np.sin(M*2*PI*(J+1)/P)
    SAM = SAM+valor[J]*np.cos(M*2*PI*(J+1)/P)

A = 2*SUM/N
B = 2*SAM/N

XS = np.zeros(N)
FIL = np.zeros(N)


for J in np.arange(N): 
    XS[J] = A*np.sin(2*PI*M*J/P)+B*np.cos(2*PI*M*J/P)


for I in np.arange(N): 
    FIL[I]= valor[I]-XS[I]

    
resultado = pd.DataFrame({'original':valor,
             'armonico':XS,
             'filtrado':FIL})
```

```python
plt.figure(figsize=(20,10))
plt.plot(resultado)
plt.legend(['original','armonico','filtrado'])
plt.grid()
```
