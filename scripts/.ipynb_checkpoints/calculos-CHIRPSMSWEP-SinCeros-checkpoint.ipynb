```python
import pandas as pd
import numpy as np
import netCDF4 as nc
import xarray as xr
import os
```

```python
#path de los archivos
path = f'{workspace}tesis/datos/procesados/'
chirps = path + 'chirps20072020.nc'
mswep = path + 'mswep20072020.nc'
```

```python
#abro mis archivos y los guardo en una variable
ds_chirps = xr.open_dataset(chirps, engine='netcdf4')
ds_mswep = xr.open_dataset(mswep, engine='netcdf4')
```

```python
#muestro CHIRPS
ds_chirps
```

```python
#muestro MSWEP
ds_mswep.where(ds_mswep.precipitation > 0.01)
```

```python
#me quedo con los valores mayores que 0.01 (trazas)
ds_chirpsSinTrazas = ds_chirps.where(ds_chirps.precip > 0.01)
ds_chirpsSinTrazasSeason = ds_chirpsSinTrazas.precip.groupby('time.season')
ds_chirpsSinTrazasSeason
ds_chirpsSinTrazasMediaDiaria = ds_chirpsSinTrazasSeason.mean(dim='time', keep_attrs=True)
ds_chirpsSinTrazasMediaDiaria.to_netcdf(path=path+'chirpsSinTrazasSeasonMedia.nc', engine='netcdf4')
```

```python
#me quedo con los valores mayores que 0.01 (trazas)
ds_mswepSinTrazas = ds_mswep.where(ds_mswep.precipitation > 0.01)
ds_mswepSinTrazasSeason = ds_mswepSinTrazas.precipitation.groupby('time.season')
ds_mswepSinTrazasSeason
ds_mswepSinTrazasMediaDiaria = ds_mswepSinTrazasSeason.mean(dim='time', keep_attrs=True)
ds_mswepSinTrazasMediaDiaria.to_netcdf(path=path+'mswepSinTrazasSeasonMedia.nc', engine='netcdf4')
```
